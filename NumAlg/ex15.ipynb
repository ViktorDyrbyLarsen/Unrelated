```python
def myRK4(dxdt, tspan, x0, n):
    a,b = tspan
    t = np.linspace(a,b, n + 1)
    h = (b-a) / n
    h2 = h / 2
    h6 = h / 6
    x = np.zeros(n+1)
    x[0] = x0
    for i in range(n):
        K1 = dxdt(t[i], x[i])
        K2 = dxdt(t[i] + h2, x[i] + h2 * K1)
        K3 = dxdt(t[i] + h2, x[i] + h2 * K2)
        K4 = dxdt(t[i] + h, x[i] + h * K3)
        x[i+1] = x[i] + h6*(K1 + 2*(K2 + K3) + K4)

    return t, x
```

```python
dxdt = lambda t, x: -x + t + np.sin(t)
dydx = lambda t, y: -x - y
```
