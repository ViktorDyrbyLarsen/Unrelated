```python
import numpy as np
import matplotlib.pyplot as plt
```

## 13.1

```python
def myTaylorOrder2(d1x,d2x,x0,interval,N):
    
    a, b = interval
    x = np.zeros(N+1)
    t = np.linspace(a,b,N+1)
    h = (b - a) / N
    x[0] = x0
    for n in range(N):
        x[n+1] = x[n] +  h * (d1x(x[n],t[n]) + h*(1/2) * d2x(x[n],t[n]))

    return np.array(t),np.array(x)

d1x = lambda x_prev, t: x_prev + t
d2x = lambda x_prev, t: x_prev + t + 1 
x0 = 1
interval= [0, 0.5]

myTaylorOrder2(d1x,d2x,x0,interval,4)
```

## 13.2

### 1

```python
I = [0,1]
N = np.array([2**i for i in range(4,10)])
d1x = lambda x_prev, t: x_prev + t
d2x = lambda x_prev, t: x_prev + t + 1 
x0 = 1
interval= [0, 1]
data = np.zeros(len(N))
absolute_error = np.zeros(len(N))
true = 2*np.exp(1) - 1 - 1

for i, n in enumerate(N):
    data[i] = myTaylorOrder2(d1x,d2x, x0,  interval, n)[1][-1]
    absolute_error[i] = abs(data[i] - true)


print(data)
print(absolute_error*N**2)
```

### 2

```python
interval = [0,10]
N = 100

d1x = lambda x_prev, t: x_prev + t
d2x = lambda x_prev, t: x_prev + t + 1
x0 = 1 



t, x = myTaylorOrder2(d1x,d2x, x0,  interval, N)
true = np.array([2*np.exp(i) - i - 1 for i in t])

absolute_error = abs(x - true)

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.plot(t, true, 'k--', label='Exact')
plt.plot(t, x, 'b', label='Taylor order 2')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.title('Numerical vs. Exact Solution')
plt.legend()
plt.grid(True)

plt.subplot(1,2,2)
plt.plot(t, absolute_error, 'r')
plt.xlabel('t')
plt.ylabel('|x_num - x_true|')
plt.title('Absolute Error')
plt.grid(True)

plt.tight_layout()
plt.show()
```
