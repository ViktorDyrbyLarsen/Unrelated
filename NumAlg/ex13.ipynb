## Exercise 13.1

```python
def myTaylorOrder2(d1x,d2x,x0,a,b,N):
    x = np.zeros(N)
    t = np.linspace(a,b,N)
    h = (b - a) / N
    x[0] = x0
    for n in range(N-1):
        x[n+1] = x[n] +  h * (d1x(x[n],t[n]) + h*(1/2) * d2x(x[n],t[n]))

    return t,x

d1x = lambda x_prev, t: x_prev + t
d2x = lambda x_prev, t: x_prev + t + 1 
x0 = 1
a, b = 0, 0.5

myTaylorOrder2(d1x,d2x,x0,a,b,10)
```
