```python
from sympy import *

def taylor(expression, order,x0):
    x = symbols("x")
    approx = 0
    for n in range(order+1):
        approx += diff(expression,x,  n).subs(x,x0) * (x - x0)**n/ factorial(n)

    return approx
```

```python
x = symbols("x")
expr = (exp(2*x) - exp(x))
taylor(expr, 3,1)
```

```python
a = [1,2,3]
b = [4,5]
x = [10,0.3,4]
def alg(a,b,x):
    for i in range(len(a)-1):
        x[i] = a[i]/x[i]
    x[-1] = (x[-1] - sum(b[i]/a[i]*x[i] for i in range(len(b)))) / a[-1]
    return x

alg(a,b,x)
```
